```python
# Visualization 1: Total Major Crime Incidents by Year (Simulated Structured Dataset)

import pandas as pd
import matplotlib.pyplot as plt

# Creating a structured dataset similar to Toronto Crime Data (aggregated yearly totals)
data = {
    "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Total_Major_Crimes": [42000, 43500, 44800, 46000, 47200, 48900, 45000, 47000, 51000, 54500, 58000]
}

df = pd.DataFrame(data)

# Create line plot
plt.figure()
plt.plot(df["Year"], df["Total_Major_Crimes"])
plt.title("Total Major Crime Incidents in Toronto (2014â€“2024)")
plt.xlabel("Year")
plt.ylabel("Number of Incidents")
plt.grid(True)
plt.show()
```

```python
# Visualization 2: Top 10 Neighbourhoods by Auto Theft Incidents (Most Recent Year - Simulated)

import pandas as pd
import matplotlib.pyplot as plt

# Creating a structured dataset similar to Toronto Crime Data (neighbourhood auto theft counts)
neighbourhood_data = {
    "Neighbourhood": [
        "Downtown Core", "Scarborough Village", "Etobicoke North",
        "North York Centre", "Weston", "Jane-Finch",
        "Parkdale", "Malvern", "Rexdale", "Willowdale"
    ],
    "Auto_Theft_Incidents": [950, 870, 820, 790, 760, 730, 710, 690, 670, 650]
}

df_neigh = pd.DataFrame(neighbourhood_data)

# Sort descending
df_neigh = df_neigh.sort_values(by="Auto_Theft_Incidents", ascending=True)

# Create horizontal bar chart
plt.figure()
plt.barh(df_neigh["Neighbourhood"], df_neigh["Auto_Theft_Incidents"])
plt.title("Top 10 Toronto Neighbourhoods by Auto Theft Incidents (2024)")
plt.xlabel("Number of Incidents")
plt.ylabel("Neighbourhood")
plt.show()
```
